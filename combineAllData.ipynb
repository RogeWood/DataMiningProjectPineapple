```python
import pandas as pd
describe = ""
# load data
# all_df = pd.read_csv('y_image_data_allTogether_predict.csv')
side_df = pd.read_csv('y_image_data_side_predict.csv')
bottom_df = pd.read_csv('y_image_data_bottom_predict.csv')
y_df = pd.concat([side_df, bottom_df])
y_df = y_df.apply(lambda x: x.fillna(0) if x.dtype.kind in 'biufc' else x.fillna('.'))
# y_df = all_df
y_df
```

```python
# 創建一個新的 DataFrame
predict_df = pd.DataFrame(columns=['ID', 1, 2, 3, 4])

# 初始化新 DataFrame 的計數列
predict_df[[1, 2, 3, 4]] = 0
predict_df
```

```python

sideORbottomList = ["side", "bottom"]
versionList = ["12", "13", "14", "23", "24", "34", "all"]
# 遍歷每一行並計算加權計數
for index, row in y_df.iterrows():
    ID = row['ID']
    counts = {1: 0, 2: 0, 3: 0, 4: 0}

    # 計算加權計數
    for s in sideORbottomList:
        for version in versionList:
            col = f'pred_{s}_{version}'
            if row[col] != 0:
                # counts[row[col]] += 1
                if  4 == row[col]:
                    counts[row[col]] += 2
                    print(row[col])
                else:
                    counts[row[col]] += 1


    # 將結果添加到新的 DataFrame
    new_row = [ID, counts[1], counts[2], counts[3], counts[4]]
    predict_df.loc[len(predict_df)] = new_row
predict_df
```

```python
# 按ID分組，並對每一列進行求和
grouped_df = predict_df.groupby('ID').sum().reset_index()
grouped_df
```

```python
# 找到每行中值最大的列名稱
grouped_df['Classification Image'] = grouped_df[[1, 2, 3, 4]].idxmax(axis=1)
grouped_df
```

```python
# 計算每個Classification voice的出現次數
classification_counts = grouped_df['Classification Image'].value_counts()
classification_counts
```

```python
grouped_df.to_csv(f'final_{describe}.csv', index=False)
```
